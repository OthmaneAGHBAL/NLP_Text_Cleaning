```python
# Imports
import pandas as pd
import spacy
import nltk
import string

from nltk.corpus import stopwords
from cleantext import clean
from beautifultable import BeautifulTable
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer


nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')


# Read the text from a CSV file and store it in a list
df = pd.read_csv('/content/sample.csv')
text_list = df['text'].tolist()
```

```python
# Conversion to lowercase
for i in range(len(text_list)):
  text_list[i] = text_list[i].lower()

# Remove emojis
for i in range(len(text_list)):
    text_list[i] = clean(text_list[i], no_emoji=True)

# Remove Twitter mentions
for i in range(len(text_list)):
  text_list[i] = ' '.join(word for word in text_list[i].split() if not word.startswith('@'))

# Remove URLs
for i in range(len(text_list)):
  text_list[i] = ' '.join(word for word in text_list[i].split() if not word.startswith('http'))

# Remove punctuations
for i in range(len(text_list)):
  text_list[i] = ''.join(char for char in text_list[i] if char not in string.punctuation)

# Tokenization
for i in range(len(text_list)):
    text_list[i] = word_tokenize(text_list[i])

# Remove digits
for i in range(len(text_list)):
  text_list[i] = [word for word in text_list[i] if not word.isdigit()]

# Remove stop words
stop_words = set(stopwords.words('english'))
for i in range(len(text_list)):
  text_list[i] = [word for word in text_list[i] if word not in stop_words]

# Lemmatize words
lemmatizer = WordNetLemmatizer()
for i in range(len(text_list)):
  text_list[i] = [lemmatizer.lemmatize(word) for word in text_list[i]]
```

```python
# Visualization
  # ID : author_id
  # TEXT_INPUT : original text
  # TEXT_OUTPUT : the cleaned text
t = BeautifulTable()
t.column_headers = ['ID', 'TEXT_INPUT', 'TEXT_OUTPUT']

for i in range(len(text_list)):
  t.append_row([ df['author_id'][i] ,  df['text'][i],' '.join(text_list[i]) ])

print(t)
```
